### JSON Module

```python
import json
```

```python
json.__file__
```

```python
# JSON string
employee = '{"id":"09", "name": "Nitin", "department":"Finance"}'

# Convert string to Python dict
employee_dict = json.loads(employee)
print(employee_dict)
print(type(employee_dict))
print("\n")

# Convert Python dict to JSON
json_object = json.dumps(employee_dict, indent=4)
print(json_object)
print(type(json_object))
```

### Pickle

```python
#Pickling is the process of serializing an object. 
#Serializing means storing the object in the form of binary representation
```

```python
#Function used to do this is dump()
#Example:
import pickle
py_dict = { 'name': 'harry', 'salary':9000, 'language': 'Hindi' }
myfile = open('file1','wb')
pickle.dump(py_dict,myfile)
myfile.close()
```

```python
# But the file made using dump() funtion is in binary and can not be read easily and therefore the file needs to be UNPICKLED.
# This is done by using load() function, and hence the file can be read easily
```

```python
myfile = open('file1','rb')
py_dict = pickle.load(myfile)
print(py_dict)
myfile.close()
```
